```python
import pandas as pd
from sqlalchemy import create_engine


DB_USER = "postgres"
DB_PASSWORD = "Ubuntu99!"
DB_HOST = "127.0.0.1"
DB_PORT = "5432"
DB_NAME = "postgres"

conn_string = f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
engine = create_engine(conn_string)


query = "SELECT * FROM public.niilsn_data;"
niilsen_data = pd.read_sql(query, engine)

print(niilsen_data.head())
```

```python
position_counts = niilsen_data['Position'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Player Positions')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(niilsen_data['Age'], bins=20, kde=True)
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(
    x='Position', 
    y='Value', 
    data=niilsen_data,
    palette='Set2'
)
plt.title('Player Market Value by Position (2020/21 Year)')
plt.xticks(rotation=45)
plt.show()
```

```python
years = ['17/18', '18/19', '19/20', '20/21']
goals_by_year = [niilsen_data[f'Gls ({year})'].sum() for year in years]

plt.figure(figsize=(10, 6))
sns.lineplot(x=years, y=goals_by_year, marker='o', linewidth=2.5)
plt.title('Total Goals Across Years')
plt.xlabel('Year')
plt.ylabel('Total Goals')
plt.grid(True)
plt.show()
```

```python
metrics_20_21 = [
    'Gls (20/21)', 
    'Ast (20/21)', 
    'xG (20/21)', 
    'xA (20/21)', 
    'Value', 
    'Age'
]

corr_matrix = niilsen_data[metrics_20_21].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix (2020/21 Year)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='Gls (20/21)', 
    y='Ast (20/21)', 
    size='Value', 
    hue='Position',
    sizes=(20, 500),  
    alpha=0.7,
    data=niilsen_data,
    palette='tab10'
)
plt.title('Goals vs. Assists (Bubble Size = Player Value)')
plt.xlabel('Goals (20/21)')
plt.ylabel('Assists (20/21)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(
    x='Position', 
    y='Age', 
    data=niilsen_data, 
    palette='viridis',
    inner='quartile'
)
plt.title('Age Distribution by Position')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
selected_columns = [
    'Club', 'Age', 'Position', 'Nation', 'Value', 'Contract Years Left'
]

selected_columns += [col for col in niilsen_data.columns if col.endswith('20/21')]



important_metrics = ['Gls', 'Ast', 'xG', 'xA', 'Min', 'MP', 'Starts']  
for season in ['19/20', '18/19', '17/18']:
    selected_columns += [f"{metric} ({season})" for metric in important_metrics]
```

```python
analysis_df = niilsen_data[selected_columns].copy()

analysis_df['Contract Years Left'] = pd.to_numeric(
    analysis_df['Contract Years Left'], 
    errors='coerce'
)

print(f"Average remaining contract: {analysis_df['Contract Years Left'].mean(skipna=True):.1f} years")


print("="*50)
print("Basic Descriptive Statistics:")
print("="*50)

categorical_cols = ['Club', 'Position', 'Nation']
print("\nCategorical Variables:")
print(analysis_df[categorical_cols].describe(include='object'))
```

```python
numerical_cols = [col for col in selected_columns if col not in categorical_cols]
print("\nNumerical Variables:")
print(analysis_df[numerical_cols].describe().transpose())

print("\n" + "="*50)
print("Additional Statistics:")
print("="*50)

print("\nTop 10 Clubs:")
print(analysis_df['Club'].value_counts().head(10))
```

```python
print("\nPosition Distribution:")
print(analysis_df['Position'].value_counts())

print("\nTop 10 Nationalities:")
print(analysis_df['Nation'].value_counts().head(10))

print("\nContract Years Left Analysis:")
print(f"Players expiring this year: {(analysis_df['Contract Years Left'] == 0).sum()}")
print(f"Average remaining contract: {analysis_df['Contract Years Left'].mean(skipna=True):.1f} years")
```

```python

print("\nContract Years Left Analysis:")

analysis_df['Contract Years Left'] = pd.to_numeric(
    analysis_df['Contract Years Left'], 
    errors='coerce'
)

print(f"Players expiring this year: {(analysis_df['Contract Years Left'] == 0).sum()}")
print(f"Average remaining contract: {analysis_df['Contract Years Left'].mean(skipna=True):.1f} years")
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.histplot(analysis_df['Age'], bins=20, kde=True)
plt.title('Age Distribution')

plt.subplot(2, 2, 2)
sns.histplot(np.log1p(analysis_df['Value']), bins=20, kde=True)
plt.title('Log-Transformed Value Distribution')

plt.subplot(2, 2, 3)
analysis_df['Position'].value_counts().plot(kind='bar')
plt.title('Position Distribution')
```
